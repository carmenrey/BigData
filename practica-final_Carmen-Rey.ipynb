```python
from functools import reduce

'''
13. Dada una lista de Artículos con sus precio. Define las siguientes funciones:
Puedes definir más funciones si lo consideras necesario.
'''
articulos = {
    'Camisa': 20,
    'Pantalón': 30,
    'Calcetines': 5,
    'Zapatos': 50,
    'Gorra': 10,
    'Bufanda': 15,
    'Gafas': 25,
    'Reloj': 35,
    'Corbata': 40,
}

compra = ['Camisa', 'Pantalón', 'Pantalón', 'Gorra', 'Gafas', 'Corbata']


# A. Una función que calcule el precio total de la compra                      
shop_prices = [articulos[compra[i]] for i in range(len(compra))]    #Devuelve una lista con el precio de todos los artículos de la lista compra
addition = lambda x, y: x + y                                       #Devuelve la suma de dos o más elementos

total = reduce(addition, shop_prices)                               #Calcula el precio total de la compra                   

# B. Una función que calcule el precio total de la compra con un descuento del 10%
total_discount = reduce(addition, list(map(lambda x: x - x * 0.1, shop_prices)))

# C. Una función que calcule el precio total de la compra con un descuento del 10% si la compra supera los 100€
filter_disc = lambda x: x - x * 0.1 if x > 100 else x           #Filtra una cantidad: si es mayor a 100, le aplica un 10% de descuento

total_disc_filtered = filter_disc(total)                        #Calcula el precio de la compra con el descuento si esta supera los 100€

# D. Una función que calcule el precio total aplicando el IVA (21%)
total_iva = reduce(addition, list(map(lambda x: x + x * 0.21, shop_prices)))

# E. Una función que liste los artículos cuyo precio es superior a 20€
high_prices = [key for key, value in articulos.items() if value > 20]

print('Total: ', total, '\nTotal con descuento: ', total_discount, '\nTotal con descuento condicional: ', total_disc_filtered, '\nTotal con IVA: ', 
round(total_iva, 2), '\nArtículos más caros: ', high_prices)
```

```python
'''
14. Dada una lista de tuplas con el nombre de un alumno, apellidos, curso y sus notas. 

 Define una funcion que reciba el curso y saque una lista en la que aparezca el nombre y apellidos y el promedio de sus notas.
 Puedes definir más funciones si lo consideras necesario.
'''

alumnos = [('Juan', 'Pérez', '1', [2, 6, 7, 8, 9]),
            ('Ana', 'García', '2', [5, 6, 7, 7, 7]),
            ('Luis', 'González', '1', [3, 6, 7, 8, 9]),
            ('María', 'Martínez', '2', [5, 5, 7, 8, 9]),
            ('Pedro', 'Rodríguez', '1', [5, 6, 3, 8, 9]),
            ('Lucía', 'Hernández', '2', [5, 6, 4, 8, 9.5]),
            ('Marta', 'López', '1', [5, 6, 7, 8, 10]),
            ('Sara', 'Díaz', '2', [5, 5.5, 7, 7, 9]),
            ('Javier', 'Sánchez', '1', [5, 6, 7, 8, 9]),
            ('Miguel', 'Fernández', '2', [3.4, 6, 7, 8, 9]),
            ('Sergio', 'Jiménez', '1', [1, 6, 6, 8, 9]),
            ('Sandra', 'Ruiz', '2', [5, 6, 10, 8, 10]),
            ('Pablo', 'Álvarez', '1', [10, 6, 1, 8, 9]),
            ('María', 'Gómez', '2', [5, 2, 7, 9.9, 9]),
]

#Función que calcula la media de una lista de elementos
average = lambda l: reduce(lambda x, y: x + y, l)/len(l)

#Función que recibe una lista de alumnos y devuelve otra lista con los nombres, apellidos y medias
students = lambda x: [x[0], x[1], average(x[3])]

#Función que filtra la lista de alumnos en función de la variable course y devuelve la lista de alumnos que están en dicho curso
course = '1'
filter_students = list(filter(lambda alumnos: course in alumnos, [alumnos[i] for i in range(len(alumnos))]))

#Mapeo la lista de alumnos filtrados y, con la función students, obtengo su media junto con sus datos
list(map(students, filter_students))

```
